```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LinearAxis, Range1d
```

```python
data_file = "220918_midland_sky.csv"
output_notebook()
```

```python
d = np.loadtxt(data_file, delimiter=",", dtype=float)
x = d[:,0]
y = d[:,1]
v = d[:,2]

# create a new plot with a title and axis labels
p = figure(title="Flight data", x_axis_label='time', y_axis_label='altitude')

p.line(x, y, legend_label="flight", color="red", line_width=2)
p.extra_y_ranges = {"gee": Range1d(start=0, end=7)}
p.line(x, v, legend_label="gforce", color="blue", line_width=2, y_range_name="gee")
p.add_layout(LinearAxis(y_range_name="gee"), 'left')
show(p)
```
